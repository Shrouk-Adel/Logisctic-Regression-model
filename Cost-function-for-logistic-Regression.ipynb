```python
def sigmoid (z):
    # g= sigmoid(z)
    g=1/(1+np.exp(-z))
    return g
```

### cost function for logistic regression

```python
def compute_cost_logistic(x,y,w,b):
    m = x.shape[0]
    cost=0.0
    for i in range(m):
        z_i=np.dot(x[i],w)+b
        f_wb_i=sigmoid(z_i)
        cost  = cost + ( - y[i]*np.log(f_wb_i) - (1-y[i]) * np.log(1-f_wb_i))
        
    cost = cost / m
      
    return cost
        
 
```

```python
w=np.array([1,1])
b=-3

print(compute_cost_logistic(x,y,w,b))
```
